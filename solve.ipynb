```python
%load_ext lab_black
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_df = pd.read_csv("training_sample.csv")
```

```python
def plot_unit(df, unitid):
    sns.set()
    sns.set_style("whitegrid")
    fig, ax = plt.subplots(nrows=5, figsize=(10, 15), sharex=True)
    df_u = df.loc[df["unitID"] == unitid]
    sns.lineplot(x="weekID", y="outcome", data=df_u, ax=ax[0])
    sns.lineplot(x="weekID", y="treatment", data=df_u, ax=ax[1])
    sns.lineplot(x="weekID", y="X1", data=df_u, ax=ax[2])
    sns.lineplot(x="weekID", y="X2", data=df_u, ax=ax[3])
    sns.lineplot(x="weekID", y="X3", data=df_u, ax=ax[4])
```

```python
plot_unit(train_df, "UNIT04257")
```
